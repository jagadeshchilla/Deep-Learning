```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv("DailyDelhiClimateTest.csv")
```

```python
df=df[['date','meantemp']]
```

```python
df 
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
```

```python
x1=df['meantemp'].ewm(alpha=0.9).mean()
```

```python
x1
```

```python
df['ewma']=x1
```

```python
df
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
plt.plot(df['date'],x1,color='black',label='alpha=0.9')
plt.legend()
plt.show()
```

```python
x2=df['meantemp'].ewm(alpha=0.98).mean()
```

```python
x2
```

```python
df['ewma']=x2
df
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
plt.plot(df['date'],x2,color='black',label='alpha=0.98')
plt.legend()
plt.show()
```
